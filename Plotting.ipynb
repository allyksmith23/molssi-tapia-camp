```python
import numpy
distance_file = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances[0,:]
data = distances[1:,:]
data = data.astype(numpy.float)
```

```python
print(headers)
```

```python
print(data)
```

```python
import matplotlib.pyplot as, plt
```

```python
sample = headers[1]
sample2 = headers[2]

plt.plot(data[:,1], label=sample)
plt.plot(data[:,2], label=sample2)
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstom)')
plt.legend()  #need this fcn to show sample name


plt.savefig(F'{two_samples.png}', dpi=300)  #big F converts stuff to string #dpi specifies resolution
```

```python
num_columns = len(headers)
print(num_columns)
print(F'The number of columns is {num_columns}')

for col in range(1, num_columns):
    plt.figure()
    print(F'The column number is {col}')
    plt.plot(data[:,col], label=headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstom)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png', dpi=200)
```

```python
num_columns = len(headers)
print(num_columns)
print(F'The number of columns is {num_columns}')

for col in range(1, num_columns):
    print(F'The column number is {col}')
    plt.plot(data[:,col], label=headers[col])
    
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstom)')
plt.legend()
plt.savefig(F'{headers[col]}.png', dpi=200)
```
